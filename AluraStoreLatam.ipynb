### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(15)
```

#1. Análisis de facturación



```python
def facturacion_total(tienda):
  suma=0
  for i in range(len(tienda)):
    precio = tienda.loc[i,'Precio'] #loc.[indice_fila,'nombre columna'] loc.=location
    suma += precio
  return suma

print(f'TOTAL FACTURACION TIENDA 1 : ${facturacion_total(tienda)} .-')
print(f'TOTAL FACTURACION TIENDA 2 : ${facturacion_total(tienda2)} .-')
print(f'TOTAL FACTURACION TIENDA 1 : ${facturacion_total(tienda3)} .-')
print(f'TOTAL FACTURACION TIENDA 1 : ${facturacion_total(tienda4)} .-')
```

```python
def resumen_ingresos(tiendas):
    resultados = {}
    for i, tienda in enumerate(tiendas, start=1):
        nombre = f"Tienda {i}"
        total = float(facturacion_total(tienda))
        resultados[nombre] = total
    return resultados

lista_tiendas=[tienda,tienda2,tienda3,tienda4]
resumen= resumen_ingresos(lista_tiendas)
print(resumen)
```

# 2. Ventas por categoría

```python
#cantidad de productos vendidos por categoria de cada tienda
```

```python
def prod_vendidos_por_categoria(tienda):
  categorias = {}
  for i in range(len(tienda)):
    categoria_producto = tienda.loc[i,"Categoría del Producto"]
    if categoria_producto in categorias:
      categorias[categoria_producto] +=1
    else :
      categorias[categoria_producto] = 1

  return categorias

def listardict(tienda):
  for categoria, cantidad in prod_vendidos_por_categoria(tienda).items():
      print(f'\t {categoria}: {cantidad}')

print('TIENDA 1:')
listardict(tienda)
print('----------------------------------')
print('TIENDA 2:')
listardict(tienda2)
print('----------------------------------')
print('TIENDA 3:')
listardict(tienda3)
print('----------------------------------')
print('TIENDA 4:')
listardict(tienda4)

```

# 3. Calificación promedio de la tienda


```python
def calificacion_promedio_por_tienda(tienda):
  suma= 0
  cont = 0
  promedio = 0
  for i in range(len(tienda)):
    suma += tienda.loc[i,"Calificación"]
    cont += 1
  promedio = suma/cont
  return promedio

print(f"CALIFICACION TIENDA 1 : {calificacion_promedio_por_tienda(tienda)}")
print()
print(f"CALIFICACION TIENDA 2 : {calificacion_promedio_por_tienda(tienda2)}")
print()
print(f"CALIFICACION TIENDA 3 : {calificacion_promedio_por_tienda(tienda3)}")
print()
print(f"CALIFICACION TIENDA 4 : {calificacion_promedio_por_tienda(tienda4)}")
```

# 4. Productos más y menos vendidos

```python
#por tienda y en todas las sucursales
#5 productos mas vendidos y 5 menos vendidos: mostrar nombre de producto, cantidad de unidades vendidas, precio unitario y total ventas
#para esto: identificar producto que se repite mas veces
```

```python
def ventas_productos_unidad(tienda):

    cantidad_ventas = tienda["Producto"].value_counts()
    suma_ventas_producto = tienda.groupby("Producto")["Precio"].sum()
    ventas = cantidad_ventas.reset_index()
    ventas.columns = ["PRODUCTO", "CANTIDAD DE VENTAS"]
    ventas["TOTAL"] = ventas["PRODUCTO"].map(suma_ventas_producto)

    ventas_ordenado = ventas.sort_values(by="CANTIDAD DE VENTAS", ascending=False)

    return ventas_ordenado

def ventas_productos_precio(tienda):
  cantidad_ventas = tienda["Producto"].value_counts()
  suma_ventas_producto = tienda.groupby("Producto")["Precio"].sum()
  ventas = cantidad_ventas.reset_index()
  ventas.columns = ["PRODUCTO", "CANTIDAD DE VENTAS"]
  ventas["TOTAL"] = ventas["PRODUCTO"].map(suma_ventas_producto)

  ventas_ordenado = ventas.sort_values(by="TOTAL", ascending=False)

  return ventas_ordenado


def mas_vendidos(lista_productos):
  return lista_productos.head(5)

def menos_vendidos(lista_productos):
  return lista_productos.tail(5)

def ranking_por_tienda(tienda):
  lista_productos_unidad = ventas_productos_unidad(tienda)
  lista_productos_precio = ventas_productos_precio(tienda)
  mas_vendidos_unidad=mas_vendidos(lista_productos_unidad)
  menos_vendidos_unidad=menos_vendidos(lista_productos_unidad)
  mas_vendidos_precio = mas_vendidos(lista_productos_precio)
  menos_vendidos_precio = menos_vendidos(lista_productos_precio)

  print("_________VENTAS POR UNIDAD_________")
  print("MAS VENTAS POR UNIDAD")
  print(mas_vendidos_unidad)
  print("MENOS VENTAS POR UNIDAD")
  print(menos_vendidos_unidad)

  print("_________VENTAS POR PRECIO_________")
  print("MAS VENTAS POR PRECIO")
  print(mas_vendidos_precio)
  print("MENOS VENTAS POR PRECIO")
  print(menos_vendidos_precio)

#ranking_por_tienda(tienda)

lista_todas_las_sucursales = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

ranking_por_tienda(lista_todas_las_sucursales)
```

# 5. Envío promedio por tienda

```python
def envio_promedio(tienda):
  promedio = tienda["Costo de envío"].mean().round(3)

  return promedio

print(f"COSTO DE ENVIOS PROMEDIO TIENDA 1 {envio_promedio(tienda)}")
print(f"COSTO DE ENVIOS PROMEDIO TIENDA 2 {envio_promedio(tienda2)}")
print(f"COSTO DE ENVIOS PROMEDIO TIENDA 3 {envio_promedio(tienda3)}")
print(f"COSTO DE ENVIOS PROMEDIO TIENDA 4 {envio_promedio(tienda4)}")
```
